```python
import os
```

```python
avoid = ["1010s1120s2001-2s5300-1", "1010s1121s2001-2s5101-1", "1010s1121s2001-2s5101-2", "1010s1121s2001-2s5301-1",
"1010s1121s2001-2s5302-1", "1010s1121s3001-2s5001-1"]
```

```python
piglife_path = "/hd2/marcos/research/repos/pig-segmentation-distill/data/PigLife"
sets = os.listdir(piglife_path)

new_total = 0
for set in sets:
    amount = 0
    excluded = 0
    images_path = os.path.join(piglife_path, set, "images")
    for i in os.listdir(images_path):
        video_id = resultado = i.rsplit('-', 1)[0]
        if video_id in avoid:
            excluded += 1
            continue
        amount += 1
    new_total += (amount - excluded)
    print(f"Set: {set} | Total images: {amount} | Avoided: {excluded} | Left: {amount - excluded}")
```
