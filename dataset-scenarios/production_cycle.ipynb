```python
import os
```

```python
test_path = "/hd2/marcos/research/repos/pig-segmentation-distill/data/PigLife/test/images"
validation_path = "/hd2/marcos/research/repos/pig-segmentation-distill/data/PigLife/val/images"
train_path = "/hd2/marcos/research/repos/pig-segmentation-distill/data/PigLife/train/images"
dataset = [test_path, validation_path, train_path]

stages = dict()
for subset in dataset:
    images = os.listdir(subset)
    for filename in images:
        sid = filename[:4]
        if sid in stages:
            stages[sid]+= 1
        else:
            stages[sid] = 1
            
print(f"Stages: {stages}")

amount = 0
for key in stages:
    amount += stages[key]

print(f"Amount of images: {amount}")
```
