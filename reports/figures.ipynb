```python
import matplotlib.pyplot as plt

models = ['n', 's', 'm']

latencies_human = [9.29, 9.51, 15.43]
map_human = [88.1, 90.6, 91.7]

latencies_sam3 = [10.13, 10.37, 16.71]
map_sam3 = [76.7, 78.7, 79.4]

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 10

plt.figure(figsize=(7, 4.5))

plt.plot(latencies_human, map_human, 'o-', label='Human annotated', markersize=7, linewidth=1.5, color='#1f77b4')
plt.plot(latencies_sam3, map_sam3, 's--', label='SAM3 assisted', markersize=7, linewidth=1.5, color='#ff7f0e')

for i, txt in enumerate(models):
    plt.annotate(txt, (latencies_human[i], map_human[i]), xytext=(5, 2), textcoords='offset points', fontsize=9)
    plt.annotate(txt, (latencies_sam3[i], map_sam3[i]), xytext=(5, 2), textcoords='offset points', fontsize=9)

plt.xlabel('Inference Pipeline Latency (ms)', fontweight='bold')
plt.ylabel('mAP (%)', fontweight='bold')
plt.title('YOLOv8 Performance: Human vs. SAM 3 Annotation', fontsize=11, pad=15)

plt.legend(loc='lower right', frameon=True)
plt.grid(True, linestyle=':', alpha=0.6)

plt.xlim(8, 18)
plt.ylim(0, 100)

plt.tight_layout()
plt.savefig('yolo_comparison_plot.pdf')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados da Tabela II
groups = ['G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8']
n_images = [25, 74, 20, 119, 37, 42, 42, 67]
map_values = [77.2, 89.9, 90.2, 80.5, 56.7, 83.6, 89.9, 80.0]

# Criando rótulos combinados: "G1 (n=25)"
x_labels = [f'{g}\n(n={n})' for g, n in zip(groups, n_images)]

plt.figure(figsize=(8, 4.5))

color = '#1f77b4'  # Azul para os outros grupos

bars = plt.bar(x_labels, map_values, color=color, edgecolor='black', alpha=0.8)

# Adicionar o valor exato no topo de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}%', ha='center', va='bottom', fontsize=9)

plt.ylabel('mAP (%)', fontweight='bold')
plt.xlabel('Scenario groups (n = number of images)', fontweight='bold')
plt.title('Detection performance per production scenario', fontsize=11)
plt.ylim(0, 110)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('performance_by_group.pdf')
plt.show()
```
